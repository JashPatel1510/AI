One way of breaking down unstructured text is to use a bag of words technique and word vectorization.  Let's consider the set of statements below in corpus.  We want to break it down into vectors.

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()

corpus = [ 'This is the first document.', 
        'This is the second second document.',
        'And the third one.', 
        'Is this the first document?']

X = vectorizer.fit_transform(corpus)
```

If we want to see what our bag of words are:

```python
vectorizer.get_feature_names_out()
```

Converting our statements into vectors we get the following

```python
X.toarray()
```

To understand the two dimensional array above, when reading each statement, look at the vectorizer array and look at the statement.  Does the word in the array exist in the statement?

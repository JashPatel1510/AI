```python
x=np.linspace(0,5,11)  #Create a sequence from 0 to 5 with 11 values.
y=x**2  #Create a second sequence based on x**2.
```

```python
x
```

```python
y
```

```python
#Plot the graph of x and y.
plt.plot(x, y, 'r') #r is for red
#try  'r.'  'r.-'   'r--'  Also try other letters other than r.  Not all results are colours.   
plt.xlabel('x label')  #Label your diagram.
plt.ylabel('y label')
plt.title('My Title')
plt.show()  #If outside of the Jupyter notbook
```

```python
y2 = x ** 3   #Create a second sequence y2 
y2
```

```python
#Here we want to graph both the y and y2 sequence against x.
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])  #% from left, % from bottom, % width, % height
ax.plot(x, y, label='x**2')
ax.plot(x, y2, label='x**3')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('My Title')
ax.legend()
```

```python
#Generate a set of 100 records.  Display the values in a histogram.  
#By default there are 10 bins in the histogram.
from random import sample
data=sample(range(1, 1000), 100)
plt.hist(data)
```

Create a box and whisker diagram based on the data generated.  A box and whisker is broken into quartiles.  To understand quartiles we sort the data from least to greatest.  As we move through our data from least to greatest Q1 is 25%, Q2 is 50%, Q3 is 75%.  Our box edges are Q1 and Q3, while the middle line is Q2.  You can find some more info at https://en.wikipedia.org/wiki/Box_plot

```python
plt.boxplot(data)
plt.show()
```

We can also generate scatter plot graphs.  Here we generate a random set of 100 points that will be good for a linear representation.

```python
x=np.linspace(0,10,100)
y=x+0.5*np.random.randn(len(x))
plt.scatter(x, y, lw=1)
```

We can generate our data into a bar graph.  There is a difference between a bar graph and a histogram.  Remember that a bar graph is for discrete data, where a histogram is for continuous data.

```python
x=np.array([1, 2, 3, 4, 5])
y=x**2
plt.bar(x, y)
```

### What is Seaborn?
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.  Copied from: https://seaborn.pydata.org/

```python
import seaborn as sns
```

Seaborn contains some built in data sets.  You can find a collection of them here: https://github.com/mwaskom/seaborn-data <br><br>
Let's read in the build in data set for "glue.csv"

```python
glue = sns.load_dataset('glue')
```

```python
glue.head()  #Examine the first 5 records
```

```python
sns.displot(glue['Score'])  #Graph the score
```

KDE is a kernel density curve and is a way of measuring distributions over a set of data.  If you do not include the kde parameter it does not show up.

```python
sns.displot(glue['Score'], kde=True, bins=30)
```

Pairplot can be used to graph any pair of graphical data.  This can be helpful if you are looking for any trend data.

```python
sns.pairplot(glue)
```

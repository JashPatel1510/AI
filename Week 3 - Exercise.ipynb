```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
taxis = sns.load_dataset('taxis')
```

```python
taxis.head(15)
```

```python
sns.pairplot(taxis)
```

```python
x=np.linspace(0,5,11)  #Create a sequence from 0 to 5 with 11 values.
y=x**2  #Create a second sequence based on x**2.

x
y

plt.plot(x, y, 'r') #r is for red
plt.xlabel('x label')  #Label your diagram.
plt.ylabel('y label')
plt.title('My Title')
plt.show() 
```

```python
sns.barplot(data=taxis, x="fare", y="payment" )
```

```python
# A Bar graph using discrete data
sns.countplot(x="payment", data=taxis, hue="payment") 
plt.ylabel("Count of Rides")

```

```python
#A histogram using continuous data

sns.histplot(taxis['pickup'])
```

```python
#A boxplot

sns.boxplot(taxis['passengers'])
```

```python
sns.scatterplot(data=taxis, x='fare', y='total')
```

```python
sns.scatterplot(data=taxis, x='fare', y='distance')
```

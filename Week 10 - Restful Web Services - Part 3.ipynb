```python
import requests
```

While your web service from the previous notebook is running:

```python
response = requests.get("http://127.0.0.1:5000/students")
print(response.json())
```

This doesn't look like the output from postman.  We are going to use the following function to indent our response.

```python
import json

def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=True, indent=4)
    print(text)

jprint(response.json())
```

If we want to try to add a new student into our collection we start by creating a new student and calling our post request.

```python
newStu =   {"name": "Meowy", "grade": 75, "letterGrade": "B"}
response = requests.post("http://127.0.0.1:5000/students", json = newStu) 
```

```python
jprint(response.json())
```

## Natural Language Processing (NLP) 
is concerned with giving computers the ability to interpret human language.  Using speech or text data sets, NLP tries to understand contents of documents, and contextual nuances of language within them.  

### Unstructured Text
We speak in unstructured text.  Consider the phrase “Add eggs and milt to my shopping list.”
<br>
### Structured Text
&lt;Shopping List&gt; <br>
&nbsp; &nbsp; &nbsp;   &lt;item&gt;Eggs&lt;/item&gt; <br>
&nbsp; &nbsp; &nbsp;   &lt;item&gt;Milk&lt;/item&gt; <br>
&lt;/Shopping List&gt; <br>


Converting from Structured to Unstructured is Natural Language Generation (NLG) <br> <br>
Converting from Unstructured to Structured is Natural Language Understanding (NLU)

Examples of NLP:

## Components of NLP Structure:

Tokenization - Taking a string and breaking it down into chunks.

Stemming - (running, run, ran) These are all referring to the same idea, so we can group these together as run.   

Lemmatization - (Universal, University) These are not Universe.  We can learn its meaning from a dictionary definition.

Speech Tagging - Looking where a word is used in a particular sentence.  "I am going to make dinner" vs "What make is your laptop", in the first make is a verb (a word used to describe an action), in the second make is a noun (identify a person, place, or thing)

Named Entity Recognition - For a given token is there an entity associated with it. Ontario is associated with a province, where Jonathan is associated with a person's name.

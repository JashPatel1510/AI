## Numbers Example

Let’s step away from definitions and let’s consider a common example when describing Neural Networks.  Consider how a computer might recognize a handwritten number.  Write out the digits 0-9 and if you compare them with a friend, you will see variations in the shape and style of how they were written.  

 
http://neuralnetworksanddeeplearning.com/chap1.html  


As we look over the digits in the image, our eyes send images to our visual cortex.  Neurons within our brain begin to signal, and we eventually perceive the image as the representation of some numerical digit.  A neural network is meant to follow a similar idea.

Let’s assume that each box represents an 8x8 pixel grid.  This would be 64 boxes.  We could then determine if each of these boxes are white or black.  This might be our initial input into our neural network.  We would start with our first layer being 64 perceptrons, and outputting 1 if the pixel is black, or 0 if the pixel is white.  

Our next layer might be used to determine where we have adjacent coloured pixels.  Based on where these adjacent coloured pixels are the layer after can be used to determine curves and straight lines.  This process would continue until we get to our last hidden layer that would then guess which digit was written, and output that value.  

Image from: https://seralouk.medium.com/how-to-classify-handwritten-digits-using-a-multilayer-perceptron-classifier-212b7bd6e6d1

If our model was correct then we would get a true positive output, but if our model was incorrect, then the model would have to change the weights and biases of the perceptrons to try and train a more accurate model.  The more times data is processed by the neural network, the more adjustments are made to the weights and biases, and the more accurate the model becomes.

At this point, hopefully you have a rough idea of how a Neural Network works.  From here there are lots of possible types of models you can train for different types of projects.  If you want to keep exploring Neural Networks I would suggest starting with TensorFlow.

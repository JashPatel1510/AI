## Experimenting with Regex
Consider customer support handled by a chatbot. (hubspot.com) It might ask for information from you like your phone number and email which it would have to extract.  Consider the 3 following responses to the chatbot:
<br><br>
This is what you are looking for: 905-459-7533,   jonathan.penava@sheridancollege.ca

My email is:jonathan.penava@sheridancollege.ca   My phone number is: (905)-459-7533

9054597533:jonathan.penava@sheridancollege.ca

We can use the following website set to Python to try and build regular expression strings.<br>
https://regex101.com/

Using the rules of regular expression, try to build a pattern that matches the phone number and a different pattern to match the email.

You may have come up with a different pattern then me.  The result that I came up with is as follows:
<br><br>
Phone Number () isolates groups, \d{n} represents a sequence of n digits, .* represents wildcard values <br>
`(\d{3}).*(\d{3}).*(\d{4})` <br>
For an email \w* represents possible value sequence, \. represents a literal period  <br>
`(\w*\.?\w*@\w*\.\w*)` <br><br>
Trying this out with some Python code.

```python
import re

#Return a list containing every occurrence of "ai":
samples = ['This is what you are looking for: 905-459-7531,   jonathan.penava1@sheridancollege.ca',
            'My email is:jonathan.penava2@sheridancollege.ca   My phone number is: (905)-459-7532',
            '9054597533:jonathan.penava3@sheridancollege.ca']
for txt in samples:
    x = re.search("(\d{3}).*(\d{3}).*(\d{4})", txt)
    print("("+x.group(1)+")"+x.group(2)+"-"+x.group(3))
    x = re.search("(\w*\.?\w*@\w*\.\w*)", txt)
    print(x.group())
```

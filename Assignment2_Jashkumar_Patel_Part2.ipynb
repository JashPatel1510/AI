```python

import requests
import json
```

```python
# BASE = "http://127.0.0.1:5000/housing"
```

```python
    
def jprint(obj):
    print(json.dumps(obj, indent=4))
```

```python

url_sample = "http://127.0.0.1:5000/housing/sample"
response = requests.get(url_sample)
sample = response.json()

print("----- SAMPLE RESPONSE -----")
jprint(sample)

```

```python

url_explain = "http://127.0.0.1:5000/housing/explain"
response = requests.get(url_explain)

print("----- EXPLAIN RESPONSE -----")
print(response.text)
```

```python
url_eval = "http://127.0.0.1:5000/housing/evaluate"

print("----- TEST 1: ORIGINAL SAMPLE -----")
response = requests.post(url_eval, json=sample)
jprint(response.json())
```

```python
test2 = sample.copy()
first_key = list(test2.keys())[0]
test2[first_key] = test2[first_key] * 1.1

print("----- TEST 2: INCREASE FIRST FIELD BY 10% -----")
response = requests.post(url_eval, json=test2)
jprint(response.json())
```

```python
test3 = sample.copy()
test3[first_key] = test3[first_key] * 0.5

print("----- TEST 3: DECREASE FIRST FIELD BY 50% -----")
response = requests.post(url_eval, json=test3)
jprint(response.json())
```

```python
test4 = sample.copy()
for col in test4:
    test4[col] = test4[col] + 5

print("----- TEST 4: ADD 5 TO ALL FIELDS -----")
response = requests.post(url_eval, json=test4)
jprint(response.json())
```

```python
test5 = sample.copy()
for col in test5:
    test5[col] = test5[col] * 2

print("----- TEST 5: DOUBLE ALL FIELDS -----")
response = requests.post(url_eval, json=test5)
jprint(response.json())
```

## Perceptron

A Neural Network is a collection of interconnected nodes we refer to as Neurons.  These neurons send signals to each other, and linked together can perform complex tasks.  This is true for both biological and artificial neural networks.

We will eventually look at a neural network, but let’s start with a single neuron.  This type of model is also called a Perceptron; however, Neuron and Perceptron are often used interchangeably.  

Inputs: There is some collection of inputs X that are passed into a neuron.  <br><br>
Weights: The neuron takes the given input X and multiplies it by some weight.  The larger the weight the more important the input X represents to the neuron.    <br><br>
Biases: We can then add a bias to our X*W.  This is useful if the input value X is 0, or if the X*W must be manipulated further.<br><br>
Sum: Once (X*W+b) is calculated for each input, the sum is taken.<br><br>
Activation Function: The activation function is applied to the sum of (X*W+b).  You can go into all the different types of activation functions on the Wikipedia page: https://en.wikipedia.org/wiki/Activation_function 


Some of the more common types of activation functions are:
- The binary step function, which is also known as a strong function, which outputs either 0 or 1.
- The Sigmoid function, which also produces a value 0 to 1, but is considered to be more of a soft function as it produces a decimal value instead of an integer.
- The Hyperbolic Tangent, which is like the Sigmoid but produces a value in the range -1 to 1
- The Rectified Linear Unit (ReLU), which is the most commonly used for its efficiency, and there are various variations to ReLU that are also used and listed on the Wiki page.


The output of our activation function produces the output of our Perceptron.  The output of this Perceptron then feeds into one of the inputs of another Perceptron, and this process continues to produce a neural network. <br><br>
At this point you might start to be visualizing what a Perceptron might look like if we tried to code it:


```python
def perceptron(x1, x2): #Our perceptron takes 2 inputs
    w1, w2, b = 1, 1, 1 #We define our weights and biases
    sum = x1*w1 + x2*w2 + b #We take the sum of our X*W and add our bias
    output = 1/(1+numpy.exp(-sum)) #Apply our sigmoid activation function
    return output #Produce our output
```

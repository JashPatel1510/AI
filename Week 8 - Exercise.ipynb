In this exercise you are going to analyze a yelp review, and try to determine the number of stars were given.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

yelp = pd.read_csv('yelp.csv')
```

Read in the first 5 records.

```python
yelp.head()
```

Check the info of yelp

```python
yelp.info()
```

Using CountVectorizor make the X the text, and make y the stars.  You will need to do a fit_transform on X.

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(yelp['text'])
y = yelp['stars']
```

Do a train test split and a Naive Bayes on your data.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


from sklearn.naive_bayes import MultinomialNB


nb = MultinomialNB()

nb.fit(X_train,y_train)
```

Determine the accuracy of your model.

```python
predictions = nb.predict(X_test)
nb.score(X_test, y_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report

print(classification_report(y_test,predictions))
```

```python
y_pred_1or5 = nb_1or5.predict(X_test_1or5)
```

### Web Services
Web Services are components to larger applications and provide a way for one application to speak to another.  They do so using open protocols which are regulated by the World Wide Web Consortium (W3C).  Because these are standardized protocol systems, they do not belong to any specific platform or programming language.  That means that you can write a program in one language, to interact with a server running a web service in another language. <br><br>
A web service is basically a method that is accessible through a network.  They are self-contained, self describing, modular applications that can be published and invoked across the web.  Much like a web page returns HTML, a request for a web service returns data.  They are designed to be consumed internally so they do not need or contain a GUI component.  While they do not have a set user interface, many IDE's have explorers and clients built in for testing. 


We are going to build a simple one to test it out.  Start by installing Flask into your Anaconda distribution.

```python
from flask import Flask
app = Flask(__name__)
```

```python
@app.route('/')
def hello_world():
    return 'Hello, World!'
```

```python
if __name__ == '__main__':
    app.run()
```

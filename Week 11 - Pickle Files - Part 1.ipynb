We have spent a lot of time learing how to train a model.  But what if we want to use our model in another program, or if we want to use someone else's model?  We are going to make use of Pickle in this lesson, but first, let's quickly train a model.  This is just Week 4 Part 4.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')

titanic.drop('embark_town', axis=1, inplace=True)
titanic.drop('embarked', axis=1, inplace=True)
titanic.drop('deck', axis=1, inplace=True)
titanic.drop('class', axis=1, inplace=True)
titanic.drop('alive', axis=1, inplace=True)

titanic.replace('male', 0, inplace=True)
titanic.replace('female', 1, inplace=True)
titanic.replace('man', 0, inplace=True)
titanic.replace('woman', 1, inplace=True)
titanic.replace('child', 2, inplace=True)

titanic.dropna(axis=0, inplace=True)

X, y = titanic.drop('survived', axis=1), titanic['survived'].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=87)

from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(max_iter=1000)
logmodel.fit(X_train,y_train)
```

Now that we have our trained model, we want to package it in a file to be able to share it.  
Python's Pickle module allows us to store our model into a pkl file.  

This file format is native to Python which means:
- Advantage: We can use it to save and retrieve our model in another program with very little effort.
- Disadvantage: We won't be able to use it in other programming languages.

```python
import pickle
```

```python
#Identify a file we are going to create.  Name your file and end it with a .pkl
#We open a file to write, wb stands for write and binary
#We store the newly opened file as a variable
titanic_logistic_model = open("titanic_logistic_model.pkl", "wb")
```

```python
#dump will write our model into our file (the data we want to write, the file we are writing to)
pickle.dump(logmodel, titanic_logistic_model)
```

```python
#Close the file
titanic_logistic_model.close()
```
